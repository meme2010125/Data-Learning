```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #We need this to split the data

fifa = pd.read_csv('fifa.csv',)
def fix(x):
    x = str(x)
    if('+' in x):
        calc = x.split('+')
        return int(calc[0]) + int(calc[1])
    elif('-' in x):
        calc = x.split('-')
        return int(calc[0]) + int(calc[1])
    elif x.isdigit():
        return int(x)
    else:
         return int(0)
for column in fifa.iloc[:,12:73]:
    print(column, fifa[column].dtype)
    fifa[column] = fifa[column].apply(fix)
   #calculate weird numbers in rows about player performance such as 72+9 etc.

df = fifa[["Acceleration", "Free kick accuracy", "GK diving", "Long passing", "Sliding tackle", "Volleys", "Preferred Positions"]]
df = df.dropna() #get rid of rows with empty cells
df.head()
```

```python
# default parameters pandas Series.str.split() function
df['Preferred Positions'] = df['Preferred Positions'].astype('str') #calculating numbers turned column type into object, had to change it back to string but the data is gone
df['first'] = df['Preferred Positions'].str.split(' ').str[0] #splitting and keeping only 1st value 
final = df[["Acceleration", "Free kick accuracy", "GK diving", "Long passing", "Sliding tackle", "Volleys", "first"]]
final.head()
```

```python
dummies = pd.get_dummies(final["first"])
final= pd.concat([final, dummies], axis = 1) #creating dummies and assigning to different positions
final.head()
final
```
